```r
library(class)  ## other knn implementations could be faster
```

```r
n <- 10**7
p <- 10
n_new <- 1000

X <- matrix(runif(n*p), n, p)
X_new <- matrix(runif(n_new*p), n_new, p)
y <- rbinom(n, 1, 0.5)

system.time( knn(X, X_new, y, k = 13) )[3]
```

```r
knn_run_time <- function(n, p = 10, k = 13) {
    n_new <- 1000
    X <- matrix(runif(n * p), n, p)
    X_new <- matrix(runif(n_new * p), n_new, p)
    y <- rbinom(n, 1, 0.5)
    system.time(knn(X, X_new, y, k))[3]
}
```

```r
n_vec <- 10**seq(2, 7)
time_vec <- sapply(n_vec, knn_run_time)
```

```r
cbind(n_vec, time_vec)
```

```r
plot(n_vec, time_vec, type="b")
```

```r
n_vec <- 10**seq(2, 6)
time_vec <- sapply(n_vec, function(x) knn_run_time(x, p=100, k=50)  )
plot(n_vec, time_vec, type="b")
```
